---
The generalization error of a model can be express as a sum of 3 different errors

1. Bias
> Wrong assumptions about data, for example assume linearity data when they are quadratic. High Bias can result in *underfit*.
Refers to the error that is introduced by approximating a real-life problem.

2. Variance
> Excess of sensibility to small variantion in train data. High Variance can result in *overfit*. 
Refers to the amount by which the predicted model would change if we estimated it using a different training data set

3. Irreducible Error - Bayes Error

> Noise in data.
> Bayes error rate is the lowest possible error rate for any classifier of a random outcome

***Tradeoff between Bias/Variance***

Increasing the complexity of the model will usually increase its variance and reduce bias.

Reducing the complexity of the model increases its bias and reduces its variance.
